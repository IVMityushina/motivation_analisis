```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('Инструмент оценки сотрудников .csv')
```

```python
т=df['Оцените вашу нагрузку по задачам в этом месяце ']
dff = т.to_frame()
encoder = OrdinalEncoder(categories=[['Низкая', 'Нормальная', 'Высокая']])
ordinal_encoded = encoder.fit_transform(dff[['Оцените вашу нагрузку по задачам в этом месяце ']])
df['Оцените вашу нагрузку по задачам в этом месяце '] = ordinal_encoded
```

```python
dummies = df["Оцените ваши задачи в этом месяце"].str.get_dummies(sep=";")
if 'Они не соответствуют моему уровню' not in dummies:
    dummies['Они не соответствуют моему уровню']= ""

if 'Они однообразные/скучные' not in dummies:
    dummies['Они однообразные/скучные']= ""

if 'Они развивают меня, мне интересны' not in dummies:
    dummies['Они развивают меня, мне интересны']= ""

if 'Я чувствую свою ценность для проекта, выполняя их' not in dummies:
    dummies['Я чувствую свою ценность для проекта, выполняя их']= ""

df = pd.concat([df, dummies], axis=1)
```

```python
т=df['Соблюдаете ли вы режим дня?']
dff = т.to_frame()
encoder = OrdinalEncoder(categories=[['Нет', 'Частично', 'Да']])
ordinal_encoded = encoder.fit_transform(dff[['Соблюдаете ли вы режим дня?']])
df['Соблюдаете ли вы режим дня?'] = ordinal_encoded
```

```python
dummies = df["Как вы отдыхаете (во время и после работы)"].str.get_dummies(sep=";")
if 'Занимаюсь спортом' not in dummies:
    dummies['Занимаюсь спортом']= ""

if 'Смотрю ролики в интернете' not in dummies:
    dummies['Смотрю ролики в интернете']= ""

if 'Смотрю соц сети, переписываюсь' not in dummies:
    dummies['Смотрю соц сети, переписываюсь']= ""

if 'Хожу на прогулки' not in dummies:
    dummies['Хожу на прогулкии']= ""

df = pd.concat([df, dummies], axis=1)
```

```python
т=df['Опишите ваши отношения в команде за последний месяц']
dff = т.to_frame()
encoder = OrdinalEncoder(categories=[['Не взаимодействую со своей командой', 'Нормальные', 'Были конфликтные/неприятные ситуации']])
ordinal_encoded = encoder.fit_transform(dff[['Опишите ваши отношения в команде за последний месяц']])
df['Опишите ваши отношения в команде за последний месяц'] = ordinal_encoded
```

```python
т=df['Если у меня возникают проблемы на работе, то я:']
dff = т.to_frame()
encoder = OrdinalEncoder(categories=[['Не обращаюсь за помощью', 'Не знаю к кому обратиться', 'Обращаюсь, но мне не помогают', 'Знаю к кому обратиться']])
ordinal_encoded = encoder.fit_transform(dff[['Если у меня возникают проблемы на работе, то я:']])
df['Если у меня возникают проблемы на работе, то я:'] = ordinal_encoded
```

```python
dummies = df["Сталкивались ли вы за последний месяц с:"].str.get_dummies(sep=";")
if 'Нет' not in dummies:
    dummies['Нет']= ""

if 'Личные проблемы' not in dummies:
    dummies['Личные проблемы']= ""

if 'Ощущение выгорания' not in dummies:
    dummies['Ощущение выгорания']= ""

if 'Снижение мотивации' not in dummies:
    dummies['Снижение мотивации']= ""

if 'Усталость от работы' not in dummies:
    dummies['Усталость от работы']= ""


df = pd.concat([df, dummies], axis=1)
```

```python
dummies = df["Опишите взаимоотношение с вашим руководителем"].str.get_dummies(sep=";")

if 'Дружеские' not in dummies:
    dummies['Дружеские']= ""

if 'Помогающие' not in dummies:
    dummies['Помогающие']= ""

if 'Формальные' not in dummies:
    dummies['Формальные']= ""

df = pd.concat([df, dummies], axis=1)
```

```python
т=df['Довольно ли вы количеством общения с руководителем?']
dff = т.to_frame()
encoder = OrdinalEncoder(categories=[['Хотелось бы меньше', 'Хотелось бы больше', 'Да']])
ordinal_encoded = encoder.fit_transform(dff[['Довольно ли вы количеством общения с руководителем?']])
df['Довольно ли вы количеством общения с руководителем?'] = ordinal_encoded
```

```python
dummies = df["Опишите своего руководителя"].str.get_dummies(sep=";")
if 'Понимающий' not in dummies:
    dummies['Понимающий']= ""

if 'Равнодушный' not in dummies:
    dummies['Равнодушный']= ""

if 'Токсичный' not in dummies:
    dummies['Токсичный']= ""

if 'Дружелюбный' not in dummies:
    dummies['Дружелюбный']= ""

if 'Помогающий' not in dummies:
    dummies['Помогающий']= ""

if 'Вовлеченный' not in dummies:
    dummies['Вовлеченный']= ""

if 'Поддерживающий' not in dummies:
    dummies['Поддерживающий']= ""

df = pd.concat([df, dummies], axis=1)
```

```python
df.rename(columns={'Комфортно ли вам работать из дома?\n1 - совсем не комфортно, 3 - нормально, 5 - очень комфортно': 'Комфортно ли вам работать из дома?'}, inplace=True)
```

```python
df.drop(['Оцените ваши задачи в этом месяце', 'Как вы отдыхаете (во время и после работы)','Сталкивались ли вы за последний месяц с:'], axis=1 )
```

```python
new_data = {'Имя': df['Ваше имя '],
                'Уровень мотивации': df['Оцените свою мотивацию:'],
                'Приоритетность': "",
                'Организация работы': ['Все хорошо' if (x['Сколько раз вы работали на выходных за последний месяц?'] < 2 and x['Комфортно ли вам работать из дома?'] > 2 and (x['Занимаюсь спортом'] ==1 or x['Хожу на прогулки'] ==1)) else 'Есть проблемы' for _, x in df.iterrows()],
                'Рабочая задача': ['Все хорошо' if (x['Оцените вашу нагрузку по задачам в этом месяце ']==1 and x['Они однообразные/скучные']!= 1 and x['Они не соответствуют моему уровню']!= 1 and x['Если у меня возникают проблемы на работе, то я:'] ==3) else 'Есть проблемы' for _, x in df.iterrows()],
                'Эмоциональное состояние': ['Все хорошо' if ((x['Оцените ваше эмоциональное состояние:'] > 2 and x['Опишите ваши отношения в команде за последний месяц']==1 and x['Нет']==1) or x['Оцените ваше эмоциональное состояние:'] > 2) else 'Есть проблемы' for _, x in df.iterrows()],
                'Отношения с руководителем': ['Все хорошо' if ((x['Помогающие'] ==1) or x['Дружеские'] ==1) and x['Довольно ли вы количеством общения с руководителем?'] ==2 and (x['Равнодушный'] !=1 or x['Токсичный'] !=1) else 'Есть проблемы' for _, x in df.iterrows()]}
new_df = pd.DataFrame(new_data)
```

```python
def get_priority(row):
    good_conditions = 0
    for col in ['Организация работы', 'Рабочая задача', 'Эмоциональное состояние', 'Отношения с руководителем']:
        if row[col] == 'Все хорошо':
            good_conditions += 1
    if good_conditions == 4:
        return 'Все нормально'
    elif good_conditions == 3:
        return 'Требует наблюдения'
    elif good_conditions == 2:
        return 'Требует вмешательства'
    else:
        return 'Требует срочного вмешательства'

new_df['Приоритетность'] = new_df.apply(get_priority, axis=1)
```

```python
def get_recommendations(row):
    recommendations = []
    num = 1
    if row['Оцените вашу нагрузку по задачам в этом месяце '] == 0:
        recommendations.append(str(num)+'. Сотруднику не хватает задач, необходимо обсудить возможные задачи\n')
        num += 1
    if row['Оцените вашу нагрузку по задачам в этом месяце '] == 2 and row['Сколько раз вы работали на выходных за последний месяц?']>1:
        recommendations.append(str(num)+'. Сотрудник перегружен и работает в выходные, необходимо рассмотреть возможности по снижению нагрузки\n')
        num += 1
    else:
        if row['Оцените вашу нагрузку по задачам в этом месяце '] == 2:
            recommendations.append(str(num)+'. Сотрудник перегружен, необходимо обсудить с сотрудником\n')   
            num += 1
        else:
            if row['Сколько раз вы работали на выходных за последний месяц?']> 1:
                recommendations.append(str(num)+'. Сотрудник много работает на выходных, необходимо выявить причины\n')
                num += 1
   
    if row['Они не соответствуют моему уровню'] ==1 or row['Они однообразные/скучные']==1:
        recommendations.append(str(num)+'. Сотрудник не доволен своими задачами, необходимо обсудить и подобрать новые задачи\n')
        num += 1
    
    if row['Комфортно ли вам работать из дома?']<3 and row['Соблюдаете ли вы режим дня?']==0 and row['Занимаюсь спортом']==0 and row['Хожу на прогулки']==0:
        recommendations.append(str(num)+'. У сотрудника проблемы с организацией труда, необходимо аккуратно обсудить подход к организации труда с сотрудником, предложить перейти на гибрид\n')
        num += 1
    else:
        if row['Занимаюсь спортом']==0 and row['Хожу на прогулки']==0:
            recommendations.append(str(num)+'. Сотрудник не занимается спортом, необходимо намекнуть на важность физических нагрузок для поддержания в тонусе когнитивных способностей\n')
            num += 1
    
    if row['Опишите ваши отношения в команде за последний месяц']!=1 and row['Если у меня возникают проблемы на работе, то я:']!=3:
        recommendations.append(str(num)+'. Вероятно сотрудник испытывает проблемы со своей командой и не может/ не хочет получить помощь\n')
        num += 1
    else:
        if row['Опишите ваши отношения в команде за последний месяц']!=1:
            recommendations.append(str(num)+'. Сотрудник испытывает проблемы с коллективом, необходимо обсудить с сотрудником\n')
            num += 1
        else:
            if row['Если у меня возникают проблемы на работе, то я:']!=3:
                recommendations.append(str(num)+'. Сотрудник имеет проблемы с обращением за помощью, необходимо обсудить с сотрудником\n')
                num += 1
   
    if row['Оцените ваше эмоциональное состояние:']<3 and (row['Ощущение выгорания']==1 or row['Личные проблемы']==1 or row['Снижение мотивации']==1):
        recommendations.append(str(num)+'. Сотрудник испытывает эмоциональные сложности, необходимо аккуратно неформально поговорить\n')
        num += 1
    if row['Формальные']==1 and row['Дружеские']==0 and row['Помогающие']==0 and row['Довольно ли вы количеством общения с руководителем?']==1 and (row['Равнодушный'] or row['Токсичный']):
        recommendations.append(str(num)+'. Сотруднику не хватает общения со своим руководителем, вероятно у него есть темы, которые нужно с вами обсудить\n')
        num += 1
    else:
        if row['Довольно ли вы количеством общения с руководителем?']==1:
            recommendations.append(str(num)+'. Сотруднику не хватает общения с руководителем, необходимо с ним поговорить\n')
            num += 1
        else:
            if row['Довольно ли вы количеством общения с руководителем?']==0:
                recommendations.append(str(num)+'. Сотрудник испытывает избыток общения с руководителем, возможно стоит обсудить это с сотрудником и сократить колиечство дополнительных встреч\n')
                num += 1
    if not recommendations:
        recommendations.append('Рекомендаций нет')
    return "\n".join(recommendations)

new_df['Рекомендации'] = df.apply(get_recommendations, axis=1)
```

```python
new_df['Ответы от сотрудников']=df['Напишите здесь, если вы хотели бы дополнить ответы выше ']
```
